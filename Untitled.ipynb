```python
# imports
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import sklearn
```

```python
# import and create datafram 
df = pd.read_csv(r'data.csv')
df.head()
```

```python
# drop unnescessary colums like employerrId, orighiredate_key, recorddate_key, birthdate_key,terminationdate_key, 
# store_name,gener_full. termreason_desc, termtype_desc , 
df = df.drop(['recorddate_key', 
              'birthdate_key', 
              'orighiredate_key', 
              'terminationdate_key', 
              'gender_full', 'store_name', 
              'termreason_desc', 
              'termtype_desc',
              'EmployeeID'],
             axis = 1
            )
df.head()
```

```python
# we shall encode the city_names and the department_name using a label encoder due to their large number ans since their order is not relevant
# we shall encode the job_title using the ordinal encoder since rank matters in ordicnal encoders
# the bussiness_unit and gender_short shall be encoded using one hot encoder
```

```python
from sklearn.preprocessing import OrdinalEncoder
Oencoder = OrdinalEncoder()

#now, we split the dataset into training and testing dataset
X = df.drop(['STATUS'], axis =1 )
y =  df['STATUS']

from sklearn.model_selection import train_test_split
X_train, X_test , y_train, y_test =  train_test_split(X, y ,test_size = 0.25 , random_state =  42)
# spltting the into 75% training and 25% test
```

```python
# Define categorical and numerical columns

categorical_columns = ['city_name', 'department_name', 'job_title', 'gender_short', 'BUSINESS_UNIT']
numerical_columns = ['age', 'length_of_service']
categorical_columns 
```

```python
# Create preprocessing steps
from sklearn.pipeline import Pipeline
from sklearn.impute import SimpleImputer
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import LabelEncoder, OneHotEncoder

categorical_transformer = Pipeline(steps=[
    ('imputer', SimpleImputer(strategy='constant', fill_value='unknown')),
    ('onehot', OneHotEncoder(handle_unknown='ignore'))
])

numerical_transformer = Pipeline(steps=[
    ('imputer', SimpleImputer(strategy='mean'))
])
```

```python
# Combine preprocessing steps
preprocessor = ColumnTransformer(
    transformers=[
        ('num', numerical_transformer, numerical_columns),
        ('cat', categorical_transformer, categorical_columns)
    ])
```

```python
from sklearn.ensemble import RandomForestClassifier

# Create a preprocessing and modeling pipeline
model = Pipeline(steps=[('preprocessor', preprocessor),
                        ('classifier', RandomForestClassifier(n_estimators=100, min_samples_split=3, max_depth=15, random_state=42))])

```

```python
# Fit the model
model.fit(X_train, y_train)
```

```python
# Evaluate the model
train_accuracy = model.score(X_train, y_train)
test_accuracy = model.score(X_test, y_test)
```

```python
print(f"Training Accuracy: {train_accuracy:.4f}")
print(f"Testing Accuracy: {test_accuracy:.4f}")
```

```python
from sklearn.metrics import classification_report

# Print classification report
y_pred = model.predict(X_test)
print("\nClassification Report:")
print(classification_report(y_test, y_pred))
```

```python
# Save the model
import joblib
joblib.dump(model, 'random_forest_model.joblib')
print("\nModel saved in random_forest_model.joblib")
```
